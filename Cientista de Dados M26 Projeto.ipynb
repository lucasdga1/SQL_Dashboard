Abaixo temos a configuração do ambiente SQL:

```python
import sqlite3
import pandas as pd
```

```python
df_transacoes = pd.read_csv("TB_TRANSACOES_PROJETO_ECOMM.csv", delimiter=';')
df_clientes = pd.read_csv("TB_CLIENTES_PROJETO_ECOMM.csv", delimiter=';')
```

```python
conn = sqlite3.connect('projeto.db')
# Carregar o DataFrame no banco de dados SQLite - criando tb_transacoes e tb_clientes
df_transacoes.to_sql('tb_transacoes', conn, index=False, if_exists='replace')
df_clientes.to_sql('tb_clientes', conn, index=False, if_exists='replace')
```

```python
# Função para executar consultas SQL e retornar o resultado como um DataFrame
def run_query(query):
    return pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
*
FROM tb_clientes
"""
result_df = run_query(query)
result_df
```

```python
query = """
SELECT
*
FROM tb_transacoes
"""
result_df = run_query(query)
result_df
```

```python
query = """
SELECT
*
FROM tb_transacoes
LEFT JOIN tb_clientes
ON tb_transacoes.id_client = tb_clientes.id_client
"""
result_df = run_query(query)
result_df
```

```python
result_df.info()
```

```python
# Tratando os dados nulos
for i, row in result_df.iterrows():
    if pd.isnull(row['state_name']):
        result_df.at[i, 'state_name'] = 'NI'
    if pd.isnull(row['First_name']):
        result_df.at[i, 'First_name'] = 'Nome'
    if pd.isnull(row['Gender']):
        result_df.at[i, 'Gender'] = 'NI'
    if pd.isnull(row['Job_Title']):
        result_df.at[i, 'Job_Title'] = 'NI'
    else: continue
result_df = result_df.drop(columns = ['Id_client'])
result_df
```

Justifique a escolha do JOIN:

Após observar as tabelas eu percebi que na tabela de clientes faltavam ID de alguns. Ao usar o RIGHT JOIN eu perderia 71 dados de transações, devido aos ID faltantes.
Tendo em vista o fato de a tabela ter 367 registros, eliminar 71 seria uma perda impactante na análise futura.
Logo eu tomei a decisão de usar o LEFT JOIN para manter o número original de dados, e tratar os dados nulos de maneira que fosse possível desenvolver o dashboard.
O valor "NI" significa "Não Informado", e "Nome" significa "No Name" para efeitos dessa análise.

Exportando o arquivo como CSV:
